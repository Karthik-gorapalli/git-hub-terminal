## Converting all data to float type

```python
for col in data.select_dtypes(include='int64'):  # Adjust dtype as needed, e.g., 'int32'
    data[col] = data[col].astype(float)
```

```python
data["AMH(ng/mL)"] = pd.to_numeric(data["AMH(ng/mL)"], errors='coerce')
data.fillna({'AMH(ng/mL)' : data['AMH(ng/mL)'].median()} , inplace = True)
```

```python
data.info()
```

```python
data['PCOS (Y/N)'] = data['PCOS (Y/N)'].astype(int)
```

```python
data.info()
```

## Dependent and independent variables selection

```python
y = data["PCOS (Y/N)"]
x = data.drop(["PCOS (Y/N)"], axis = 1)
```

```python
corr_matrix= data.corr(method = 'pearson')
plt.subplots(figsize=(30,10))
sns.heatmap(corr_matrix,cmap="Set3", annot = True, fmt = ".2f");
plt.title("Correlation Between Features")
plt.show()
```

## Training the model

```python
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.30 , random_state = 42)
```

```python
print("X_train",len(x_train))
print("X_test",len(x_test))
print("y_train",len(y_train))
print("y_test",len(y_test))
```

## Import logistic regrression

```python
#creat the model
lr = LogisticRegression()
lr.fit(x_train,y_train)

acc_log_train = round(lr.score(x_train, y_train)*100,2) 
acc_log_test = round(lr.score(x_test,y_test)*100,2)
print(f"Training Accuracy: {acc_log_train} %")
print(f"Training Accuracy: {acc_log_test} %")

#predict the target variable
y_prediction_values = lr.predict(x_test)
```
